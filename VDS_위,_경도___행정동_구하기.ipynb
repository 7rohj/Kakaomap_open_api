## 위도, 경도 -> 주소 변환

```python
def lat_lon_to_addr(lon,lat):
    url = 'https://dapi.kakao.com/v2/local/geo/coord2regioncode.json?x={longitude}&y={latitude}'.format(longitude=lon,latitude=lat)
    headers = {"Authorization": "KakaoAK " + 'KEY'}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address_name']
    return str(match_first)
```

```python
df = pd.read_excel("vdsid 별 위도 경도.xlsx")
```

```python
df.head(3)
```

```python
newdf = df[['경도','위도']]
newdf.head(3)
```

```python
xcode = newdf['경도'].tolist()
ycode = newdf['위도'].tolist()
```

```python
len(xcode)
```

```python
binlist = []

for i in range(len(xcode)):
  print(i,'번째',lat_lon_to_addr(xcode[i], ycode[i]))
  binlist.append(lat_lon_to_addr(xcode[i], ycode[i]))

print('----finished----')
```

```python
df['행정동 주소'] = binlist
```

```python
df.to_excel('행정동 추가.xlsx')
```
