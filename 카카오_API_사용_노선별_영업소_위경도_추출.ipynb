```python
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import requests
import pprint
import json
```

```python
import time
```

```python
def addr_to_lat_lon(addr):
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={address}'.format(address=addr)
    headers = {"Authorization": "KakaoAK " + '470f388952a8853a5085484a16c64900'}
    result = json.loads(str(requests.get(url, headers=headers).text))
    match_first = result['documents'][0]['address']
    return float(match_first['x']), float(match_first['y'])
```

```python
rr = np.array(df['주소']) # 트래픽 제한 있어서 일단은 unique 한 값들만
rr_list = rr.tolist()
```

```python
ainlist = []
binlist = []
cinlist = []
for i in rr_list:
  try:
    ainlist.append(i)
    binlist.append(addr_to_lat_lon(i))
    print(addr_to_lat_lon(i))
  except Exception:
    pass
    print(i,'Sys Error💢💥💥 must check it out')
    cinlist.append(i)
```
